```python
from types import SimpleNamespace
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-07-28/penguins.csv"
penguins = pd.read_csv(url).convert_dtypes()
```

```python
t = SimpleNamespace(**{name: name for name in penguins.columns})
```

```python
penguins.value_counts(t.island)
```

```python
penguins.value_counts(t.sex, dropna=False)
```

```python
penguins.value_counts([t.species, t.island]).reset_index(name="n")
```

```python
np.mean(penguins[t.bill_depth_mm])
```

```python
penguins.assign(species=lambda df: df[t.species].str.upper())
```

```python
op = SimpleNamespace()
```

```python
def mean(name: str):

    def inner(table: pd.DataFrame):
        return table[name].mean()

    return inner

op.mean = mean
```

```python
def upper(name: str):
    
    def inner(table: pd.DataFrame):
        return table[name].str.upper()
    
    return inner

op.upper = upper
```

```python
penguins.assign(mean_bill_depth=op.mean(t.bill_depth_mm))
```

```python
penguins.groupby([t.species, t.island], as_index=False, dropna=False).agg({t.bill_depth_mm: "mean", t.bill_length_mm: "mean"})
```

```python
penguins.assign(**{t.island: op.upper(t.island)}).value_counts(t.island)
```
